```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with CSE-Aâœ¨.txt','r', encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[ap]m\s-\s'
```

```python
messages= re.split(pattern, data)[1: ]
messages
# len(messages)
```

```python
dates=re.findall(pattern, data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})

# convert message_date type
df['message_date']=pd.to_datetime(df['message_date'],format='%d/%m/%Y, %I:%M %p - ')

df.rename(columns={'message_date':'date'}, inplace=True)

df.head()
```

```python
df['user_message'].unique()
```

```python
df.shape
```

```python
# separate users and messages
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('(?:([A-Za-z ]+):|(\+?\d{2,}\s\d{5}\s\d{5}):)\s',message)
#     if entry[1:]: #user name
# #         print(users.append(entry[0]))
#         users.append(entry[2])
#         messages.append(entry[3])
    if len(entry) > 1:  # user name or phone number found
        if entry[1]:  # user name
            users.append(entry[1])
            messages.append(entry[3])
        else:  # phone number
            users.append(entry[2])
            messages.append(entry[3])
    else:
        users.append('group_notification')
        messages.append(entry[0])
     
df['user']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)

df.head()
```

```python
period=[]
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period']= period
```
